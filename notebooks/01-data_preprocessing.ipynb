## User Rating Dataset

```python
user_rating_dfs = []

for i in os.listdir('../data'):
    if i.startswith('user'):
        user_rating_dfs.append(pd.read_csv('../data/' + i))

user_rating_df = pd.concat(user_rating_dfs)
```

```python
user_rating_df.info()
```

```python
user_rating_df
```

```python
# Join user_rating_df and book_df to get the book_id based on name
book_df_reset = book_df_eng.reset_index().rename(columns={'Id': 'book_id'})
user_rating_complete = pd.merge(user_rating_df, book_df_reset, left_on='Name', right_on='Name', how='left')[['ID', 'Name', 'book_id', 'Rating_x']]
user_rating_complete.rename(columns={'Rating_x': 'Rating'}, inplace=True)
```

```python
user_rating_complete['Rating'].unique()
```

```python
rating_to_score = {
    'it was amazing': 5,
    'really liked it': 4,
    'liked it': 3,
    'it was ok': 2,
    'did not like it': 1,
    "This user doesn't have any rating" : 0
}

user_rating_complete['Rating'] = user_rating_complete['Rating'].apply(lambda x: rating_to_score[x])
```

```python
user_rating_complete['Rating'].unique()
```

```python
user_rating_complete
```

```python
user_rating_complete.to_parquet('../data/user_rating_eng.parquet')
```
